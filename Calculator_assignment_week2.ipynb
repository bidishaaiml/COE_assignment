```python
#import the math module
import math

# Define the functions for the basic calculator operations
# Defining a funtion to add two numbers
def add(x, y):
    return x + y

# Defining a function to subtract two numbers
def subtract(x, y):
    return x - y

# Defining a function to multiply two numbers
def multiply(x, y):
    return x * y

# Defining a function to divide two numbers and returning an error message if the divisor is zero
def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error! Division by zero."

# Defining a function to calculate the power of a number
def power(base,exponent):
  return base**exponent

# Defining a function to calculate the square root of a number
def square_root(x):
  if x>=0:
    return math.sqrt(x)
  else:
    return "Error! Cannot calculate square root of a negative number."

# Displaying a welcome message
print("Welcome to My Python Calculator!")

while(True):
 # Prompting the user to choose the operation to be performed
 # 1---> Addition, 2---> Substraction, 3---> Multiplication, 4---> Division, 5---> Power, 6---> Square Root, 7---> Exit
 print("Select operation:")
 print("1. Add")
 print("2. Subtract")
 print("3. Multiply")
 print("4. Divide")
 print("5. Power")
 print("6. Square Root")
 print("7. Exit")

 #Taking the user's input to select from the available operations
 # 1---> Addition, 2---> Substraction, 3---> Multiplication, 4---> Division, 5---> Power, 6---> Square Root, 7---> Exit
 choice = (input("Enter choice(1/2/3/4/5/6/7): "))

 if choice == '7':
    print("Thank you for using the Python Calculator!")
    break
 elif choice not in ['1','2','3','4','5','6','7']:
    # Handling any invalid input
    print("Invalid input")
    continue
 else:
    try:
      # For operations other than square root and power, two numbers are needed
      if choice not in ['5', '6']:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
      elif choice == '5':
        base = float(input("Enter the base number: "))
        exponent = float(input("Enter the exponent: "))
      else:
        num1 = float(input("Enter number to find the square root: "))
    except ValueError:
      print("Invalid number input. Please enter numeric values.")
      continue

    # Performing the operation based on the user's choice
    if choice == '1':
      print(f"{num1} + {num2} = {add(num1, num2)}")
      continue

    elif choice == '2':
      print(f"{num1} - {num2} = {subtract(num1, num2)}")
      continue

    elif choice == '3':
      print(f"{num1} * {num2} = {multiply(num1, num2)}")
      continue

    elif choice == '4':
      print(f"{num1} / {num2} = {divide(num1, num2)}")
      continue

    elif choice == '5':
            print(f"{base} ^ {exponent} = {power(base, exponent)}")

    elif choice == '6':
            print(f"√{num1} = {square_root(num1)}")





```
